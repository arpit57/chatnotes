```python
import datetime

# Essay on meditation (approximately 300 words)
meditation_essay = """
Meditation is a practice where an individual uses a technique – such as mindfulness, or focusing the mind on a particular object, thought, or activity – to train attention and awareness, and achieve a mentally clear and emotionally calm and stable state. Studies have shown that meditation can significantly improve the quality of life.

Meditation primarily aims to bring about a state of tranquility and mental clarity. Unlike day-to-day activities that often leave us scattered and unfocused, meditation helps in focusing our attention and eliminating the stream of jumbled thoughts that may be crowding our mind. This process may result in enhanced physical and emotional well-being.

The benefits of meditation are extensive. It not only helps reduce stress but also improves concentration, encourages a healthy lifestyle, increases self-awareness and happiness, reduces aging, benefits cardiovascular and immune health, and contributes to an overall sense of well-being. People from all walks of life use meditation to gain peace, calm, and clarity.

There are numerous meditation techniques, including guided meditation, mantra meditation, mindfulness meditation, Qi gong, Tai chi, Yoga, Zen meditation, and transcendental meditation, among others. Each method involves different ways to relax or build internal energy or force. Regardless of the type, regular practice is key to experiencing the profound benefits of meditation.

Meditation does not require a huge investment of time, and it is accessible to everyone. A few minutes of meditation daily can lead to a profound change in how one navigates through life. By reducing stress and enhancing concentration, meditation allows individuals to take control of their lifestyle and make better decisions. In the hustle of modern life, meditation serves as a refuge of calm and clarity.
"""
```

```python
import chromadb

client = chromadb.PersistentClient(path="data")
```

```python
import  os

GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
```

```python
google_ef = chromadb.utils.embedding_functions.GoogleGenerativeAiEmbeddingFunction(api_key=GOOGLE_API_KEY)
```

```python
collection = client.create_collection(name="notes", embedding_function=google_ef)
```

```python

collection = client.get_collection(name="notes", embedding_function=google_ef)
```

```python
collection.add(
    documents=[meditation_essay],
    ids=[datetime.datetime.now().strftime('%d-%m-%Y %I:%M %p')]
)
```
